```python
def read_labels(json_path):
    with open(json_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
def convert_and_save_to_yolo(json_data, type, dataset_root='/content/dataset'):
    src_data_root = os.path.join(dataset_root, 'DocXPand-25k')
    json_data = read_labels(json_data)
    data_len = len(json_data['documents'])
    print(data_len)
    documents_dir = os.path.join(src_data_root, 'documents')
    for i in range(data_len):
        image_filename = json_data['documents'][i]['filename']
        # print(f"image_filename:{image_filename}")
        dst_label_filename = f"{image_filename.split('/')[-1].split('.')[0]}.txt"

        src_image_filepath = os.path.join(documents_dir, image_filename)
        dst_image_filepath = os.path.join(dataset_root, f'DocXPand-25k_yolo/{type}/images', image_filename.split('/')[-1])
        shutil.copy(src_image_filepath, dst_image_filepath)

        polygon_annotation = json_data['documents'][i]['annotations'][0]['position']
        x1, y1 = polygon_annotation['p1']['x'], polygon_annotation['p1']['y']
        x2, y2 = polygon_annotation['p2']['x'], polygon_annotation['p2']['y']
        x3, y3 = polygon_annotation['p3']['x'], polygon_annotation['p3']['y']
        x4, y4 = polygon_annotation['p4']['x'], polygon_annotation['p4']['y']

        class_index = 0
        polygon_points = [(x1, y1), (x2, y2), (x3, y3), (x4, y4)]

        # Convert to flat list of strings: ['100', '50', '200', '80', ...]
        flat_coords = [str(coord) for point in polygon_points for coord in point]

        # Combine class index and points
        annotation_line = f"{class_index} " + " ".join(flat_coords)

        # Write to a .txt file
        # print(f"dst_label_filename: {dst_label_filename}")
        dst_label_filepath = os.path.join(dataset_root, f'DocXPand-25k_yolo/{type}/labels/{dst_label_filename}')
        with open(dst_label_filepath, "w") as f:
            f.write(annotation_line + "\n")
```

```python
# Read the annotation file

base_filename = '0019d3ae-2d60-4e68-b326-d0922dd396de-ID_CARD_TD2_A-front'
data_type = 'train'
label_path = f'/content/dataset/DocXPand-25k_yolo/{data_type}/labels/{base_filename}.txt'
image_path = f'/content/dataset/DocXPand-25k_yolo/{data_type}/images/{base_filename}.jpg'

with open(label_path, "r") as f:
    line = f.readline().strip()

# Split the line into values
parts = line.split()

# Extract class index
class_index = int(parts[0])

# Convert remaining parts to coordinate pairs
coords = list(map(float, parts[1:]))  # or float if needed

# Group into (x, y) tuples
points = [(coords[i], coords[i+1]) for i in range(0, len(coords), 2)]

# read img
img = cv2.imread(image_path)
H, W, _ = img.shape

# Define points of the polygon
pts = np.array([[points[0][0]*W, points[0][1]*H], [points[1][0]*W, points[1][1]*H], [points[2][0]*W, points[2][1]*H], [points[3][0]*W, points[3][1]*H]], np.int32)

# Reshape for polylines: (N, 1, 2)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon (True = closed polygon)
img1 = cv2.polylines(img, [pts], isClosed=True, color=(0, 255, 0), thickness=3)

# Show the image
cv2_imshow(img1)
```
